# Import Librares

```python
import requests
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup

%matplotlib inline
```

```python
all_links = []
seen = set()
next_page = "http://scrapeme.live/shop/"   # use http instead of https

while next_page:
    r = requests.get(next_page, verify=False)   # ðŸ‘ˆ disable SSL check
    soup = BeautifulSoup(r.text, "html.parser")

    # Only product detail links (skip add-to-cart)
    anchors = soup.select("li.product a.woocommerce-LoopProduct-link")
    for a in anchors:
        href = a.get("href")
        if href:
            href = href.replace("https://", "http://")   # force http
            if href not in seen:
                seen.add(href)
                all_links.append(href)

    # Find "Next" button
    nxt = soup.select_one("a.next.page-numbers")
    next_page = nxt.get("href").replace("https://", "http://") if nxt else None

print("âœ… Total product links collected:", len(all_links))

# ðŸ”Ž Output 1: show all links, one per line
print(*all_links, sep="\n")
```
